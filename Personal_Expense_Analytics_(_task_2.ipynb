```python
import pandas as pd

df = pd.read_csv("expenses.csv")

print(df.head())
```

Convert Date **Column**

```python
df['Date'] = pd.to_datetime(df['Date'])
```

Expense Category Pie **Chart**

```python
import matplotlib.pyplot as plt

category_expense = df.groupby('Category')['Amount'].sum()

category_expense.plot(kind='pie', autopct='%1.1f%%')

plt.title("Expense Category Distribution")
plt.ylabel("")

plt.show()
```

Monthly Spending Bar **Graph**

```python
monthly_expense = df.groupby(df['Date'].dt.month)['Amount'].sum()

monthly_expense.plot(kind='bar')

plt.title("Monthly Spending")
plt.xlabel("Month")
plt.ylabel("Amount")

plt.show()
```

Payment Mode **Analysis**

```python
payment_mode = df.groupby('Payment_Mode')['Amount'].sum()

payment_mode.plot(kind='bar')

plt.title("Payment Mode Usage")
plt.xlabel("Payment Mode")
plt.ylabel("Amount")

plt.show()
```

Savings **Analysis**

```python
income = 10000

total_expense = df['Amount'].sum()

savings = income - total_expense

print("Total Expense:", total_expense)

print("Savings:", savings)
```

# **Insights**

```python
1. Bills category has higher expenses.
2. UPI is the most used payment method.
3. Monthly expenses help track spending habits.
4. Savings analysis helps manage finances.
```
